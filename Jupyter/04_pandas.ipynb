### Identificacion mas específicas de los datos nulos

```python
#Filtrando las filas que contienen los valores nulos
filas_nulas = datos[datos.isnull().any(axis=1)]

print(filas_nulas)
```

### Para eliminar los datos nulos

```python
#NO ES LO MISMO DROP QUE DROPNA (DROPNA ELIMINA A LOS NULOS, DROP ELIMINA MAS ESPECÍFICA)
data_no_nulos = datos.dropna()
data_no_nulos
```

#### Función para verificar si existe una columna con un nombre x en la lista

```python
def filtro_rpg(genres):
    return 'RPG' in genres

#filtro apply (Se deben ir cargando los dataset ya corregidos, no el raw)
datosRPG = data_no_nulos[data_no_nulos['Genres'].apply(filtro_rpg)]

#Para que el dataset  que también incluya las columnas 'Títulos' y 'Rating'
nombresRPG = datosRPG[['Title','Rating']]
nombresRPG.head(20)
```

### Información del dataframe ya filtrada
